```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./Dataset/cancer_patient_data_sets.csv')
df.head()
```

```python
### Let's check the details of the dataset
df.info()
```

```python
df.describe()
```

```python
### shape of the dataset
df.shape
```

```python
### checking for null values
df.isnull().sum()
```

```python
### drop unnecessary clms from dataset
df.drop(columns=['index', 'Patient Id'], axis=1, inplace=True)
df
```

```python
df['Level'].value_counts()
```

```python
### we need to change the level from categorical to numerical
df["Level"].replace({'High': 2, 'Medium': 1, 'Low': 0}, inplace=True)
print('Cancer Levels: ', df['Level'].unique())
```

```python
### detect outliers
### no outliers presents in this dataset 
```

```python
### make sure there are no categorical clms present
df.dtypes
```

```python
corr_metrix = df.corr()
corr_metrix['Level']*100
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(corr_metrix, annot=True, cmap='Blues')
```

```python
plt.figure(figsize = (15,7))
colors = ['blue', 'red', 'green']
plt.title("Lung Cancer Chances ")
plt.pie(df['Level'].value_counts(), explode = (0.1, 0.02, 0.02), labels = ['High', 'Medium', 'Low'], autopct = "%1.2f%%", colors = colors)
plt.legend(title = "Lung Cancer Chances", loc = "lower left")
```

```python
sns.displot(df['Level'], kde=True, color = 'red')
```

```python
### data  preprocessing is done
df.to_csv('./Dataset/processed_dataset.csv', index=False)
```
